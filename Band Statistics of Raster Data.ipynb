```python
#sys module provides information about constants, functions and methods of the Python interpreter
import sys

#Loop through all raster bands for listing band statistics.
src_ds = gdal.Open( "/Users/ArshiaSultana/Desktop/PS_tmax_10.tif" )
if src_ds is None:
    print('Unable to open INPUT.tif')
    sys.exit(1)

print("[ RASTER BAND COUNT ]: ", src_ds.RasterCount)
for band in range(src_ds.RasterCount):
    band += 1
    print("[ GETTING BAND ]: ", band)
    srcband = src_ds.GetRasterBand(band)
    if srcband is None:
        continue

    stats = srcband.GetStatistics( True, True )
    if stats is None:
        continue
    print(f"STATS: Minimum={stats[0]}, Maximum={stats[1]}, Mean={stats[2]}, StdDev={stats[3]}")
```

```python
# Creating an empty dataframe
import pandas as pd

df = pd.DataFrame(columns=['Min Temp', 'Max Temp', 'Mean Temp', 'Std Dev'])
```

```python
# Inserting values into the dataframe 


min_temp = []
max_temp = []
mean_temp = []
std_dev = []

# Iterate through all the bands and store the data into individual series

for band in range(src_ds.RasterCount):
    band += 1
    srcband = src_ds.GetRasterBand(band)

    stats = srcband.GetStatistics( True, True )
    min_temp.append(stats[0])
    max_temp.append(stats[1])
    mean_temp.append(stats[2])
    std_dev.append(stats[3])
```

```python
# Check if all the 365 values were saved
print(len(min_temp))
print(len(max_temp))
print(len(mean_temp))
print(len(std_dev))
```

```python
# Add these values to the dataframe
df['Min Temp'] =  min_temp
df['Max Temp'] = max_temp
df['Mean Temp'] = mean_temp
df['Std Dev'] = std_dev
```

```python
# save this to a csv
df.to_csv("raster_output.csv")
```
